### Visualizing data
- An intersting way to display dataframe with matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,4))
plt.plot(a1,a2,color='red')
plt.xlabel('length')
plt.ylabel('height')
plt.show()
```

```python
division = [1,2,3,4]
marks = [75,45,89,75]
avg_marks = [65,75,59,76]
width = 0.70
plt.bar(division,marks,color="blue",width = 0.3)
plt.show()
```

```python
plt.barh(division,marks,color='blue')
plt.show()
```

```python
lang = ["pytho","java","c","c++"]
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
